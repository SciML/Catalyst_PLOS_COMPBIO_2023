# Preparations

```python
import bionetgen
import matplotlib.pyplot as plt
```

```python
def plot_result(result):
    r = result[0]

    for name in r.dtype.names:
        if name != "time":
            plt.plot(r['time'], r[name], label = name)
    plt.xlabel("time")
    plt.ylabel("species (counts)")
    _ = plt.legend(frameon = False)
    return plt
```

# Benchamarks

### Multistate

```python
model_multistate_ODE = bionetgen.bngmodel('../Data/bngl_files/ODE/multistate.bngl')
result_multistate_ODE = bionetgen.run(model_multistate_ODE);
plot_result(result_multistate_ODE)
```

```python
model_multistate_Gillespie = bionetgen.bngmodel('../Data/bngl_files/Gillespie/multistate.bngl')
result_multistate_Gillespie = bionetgen.run(model_multistate_Gillespie);
plot_result(result_multistate_Gillespie)
```
